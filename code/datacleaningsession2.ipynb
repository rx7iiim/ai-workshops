```python
import pandas as pd
df=pd.read_excel('datasets\Customer Call List.xlsx')
df
```

## 1. Remove Duplicate Records\n,
    \n,
    Identify and remove duplicate customer entries to ensure data integrity.

```python
df=df.drop_duplicates()
df
```

2-normalizing the first and last name columns (rmoving unwanted carcaters)

```python
df ['First_Name']= df['First_Name'].str.strip('/._')
df['Last_Name']=df['Last_Name'].str.strip('/._')
```

3-giving all the phone numbers the same format xxx-xxx-xxx

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)

listpn =list(df['Phone_Number'])
listpn1=[]
for j in listpn:
      j=str(j)
      i=3
      while i<len(j):      
            j=j[:i]+'-'+j[i:]
            i+=4
      listpn1.append(j)

df['Phone_Number']=listpn1
df
```

4-splitting the address  column to three features to orgnize our data in a better way keeping our data consistency 

```python
df[['city','street','house']] =df['Address'].str.split(',', n=2, expand=True)
```

```python
df['Paying Customer']=df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer']=df['Paying Customer'].str.replace('No','N')
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('No','N')
```

5- dropping the unusefull features 

```python
df = df.drop(columns='Address')
df = df.drop(columns='Not_Useful_Column')
```

6-filling our missing values and nulls 

```python
df=df.replace('Na','')
df=df.replace('N/a','')
df.fillna("")
```

7-exporting our new data 

```python
df.to_excel('finaltask.xlsx', index=False)
```
